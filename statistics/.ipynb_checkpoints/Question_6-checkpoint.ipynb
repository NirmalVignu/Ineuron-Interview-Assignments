```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data
data = np.loadtxt("data.csv", delimiter=",", skiprows=1)

# Calculate the change in blood pressure
change_in_blood_pressure = data[:, 2] - data[:, 1]

# Calculate the quantiles of the change in blood pressure
quantiles = np.quantile(change_in_blood_pressure, [0.25, 0.5, 0.75])

# Calculate the quantiles of a normal distribution
normal_quantiles = np.quantile(np.random.normal(0, 1, size=len(change_in_blood_pressure)), [0.25, 0.5, 0.75])

# Plot the Q-Q plot
plt.plot(quantiles, normal_quantiles)
plt.xlabel("Quantiles of change in blood pressure")
plt.ylabel("Quantiles of normal distribution")
plt.show()
```
